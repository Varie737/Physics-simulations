#TAYLOR SERIES
The purpose of this notebook is to practice mathematical modelling in python in this case... The taylor series... obviously

```python
#Importing libraries
import numpy as np
from math import factorial
import matplotlib.pyplot as plot
from IPython.display import clear_output, display
```

Try Taylor Series for cosine using the term for f(x)=cos x at centered at x=0

```python
x = np.linspace(-10, 10, 100)
y = np.zeros(len(x))
y_cos = np.cos(x)

# Create figure and axis once outside the loop
figure, axis = plot.subplots(figsize=(8, 6))

for n in range(0, 40):
    #cos(x)
    taylor_term = (-1)**n * x**(2*n)/factorial(2*n)
    y = np.add(y, taylor_term)

    clear_output(wait=True)

    # Clear the axis for the new plot
    axis.clear()

    axis.plot(x, y, label='Taylor Approximation')
    axis.plot(x, y_cos, label='cos(x)', linestyle='--')
    axis.set_xlabel('x')
    axis.set_ylabel('Value')
    axis.set_title('Taylor Series for cos(x) - Terms: {}'.format(n+1))
    axis.legend()

    axis.set_xlim(min(x), max(x)) # Use axis.set_xlim
    axis.set_ylim(-2, 2)         # Use axis.set_ylim

    axis.grid(True)               # Use axis.grid


    display(figure) # Explicitly display the figure for animation

    plot.pause(1.0) # Increased pause to make animation more visible

print(f"Difference between Taylor approximation (last term) and cos(x): {np.mean(np.abs(y - y_cos))}")
```

Notice how the series looks more like the cosine function as more terms are added. Very Cool

try for sine

```python
x = np.linspace(-10, 10, 100)
y = np.zeros(len(x))
y_sin = np.sin(x)

# Create figure and axis once outside the loop
figure, axis = plot.subplots(figsize=(8, 6))

for n in range(0, 40):
    #sin(x)
    taylor_term = (-1)**n * x**((2*n) +1)/factorial((2*n) +1)
    y = np.add(y, taylor_term)

    clear_output(wait=True)

    # Clear the axis for the new plot
    axis.clear()

    axis.plot(x, y, label='Taylor Approximation')
    axis.plot(x, y_sin, label='sin(x)', linestyle='--')
    axis.set_xlabel('x')
    axis.set_ylabel('Value')
    axis.set_title('Taylor Series for sin(x) - Terms: {}'.format(n+1))
    axis.legend()

    axis.set_xlim(min(x), max(x)) # Use axis.set_xlim
    axis.set_ylim(-2, 2)         # Use axis.set_ylim

    axis.grid(True)               # Use axis.grid


    display(figure) # Explicitly display the figure for animation

    plot.pause(1.0) # Increased pause to make animation more visible

print(f"Difference between Taylor approximation (last term) and sin(x): {np.mean(np.abs(y - y_sin))}")
```
